```python
conn = sqlite3.connect('./data/ntsb.db')
cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table' and lower(name) in('events','aircraft','events_sequence') or 1=1;")
tablas = cursor.fetchall()

print("tabla,campos,registros")
for tabla in tablas:
    tabla_nombre = tabla[0]
    
    # columnas
    cursor.execute(f"PRAGMA table_info({tabla_nombre})")
    columnas = len(cursor.fetchall())

    # registros
    cursor.execute(f"SELECT COUNT(*) FROM {tabla_nombre}")
    registros = cursor.fetchone()[0]

    
    print(tabla_nombre+","+str(columnas)+","+str(registros))

conn.close()
```

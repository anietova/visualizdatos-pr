```python
mdb_conn = pyodbc.connect(r"Driver={Microsoft Access Driver (*.mdb, *.accdb)};DBQ=" + mdb_path)

mdb_cur = mdb_conn.cursor()

sqlite_conn = sqlite3.connect(sqlite_path)
sqlite_cur = sqlite_conn.cursor()

tables = [
    "aircraft",
    "Country",
    "ct_iaids",
    "ct_seqevt",
    "dt_aircraft",
    "dt_events",
    "dt_Flight_Crew",
    "eADMSPUB_DataDictionary",
    "engines",
    "events",
    "Events_Sequence",
    "Findings",
    "Flight_Crew",
    "flight_time",
    "injury",
    "narratives",
    "NTSB_Admin",
    "Occurrences",
    "seq_of_events",
    "states"
]

for table in tables:
    print("Procesando tabla,", table)

    mdb_cur.execute(f"SELECT * FROM {table} WHERE 1=0")
    columns = [column[0] for column in mdb_cur.description]

    col_defs = ", ".join([f'"{col}" TEXT' for col in columns])
    sqlite_cur.execute(f'DROP TABLE IF EXISTS "{table}";')
    sqlite_cur.execute(f'CREATE TABLE "{table}" ({col_defs});')

    mdb_cur.execute(f"SELECT * FROM {table}")
    rows = mdb_cur.fetchall()

    # construimos insert on tantos ? como columnas tengamos
    placeholders = ", ".join(["?"] * len(columns))
    insert_sql = f'INSERT INTO "{table}" VALUES ({placeholders})'
    # carga tipo bulk
    sqlite_cur.executemany(insert_sql, rows)
    sqlite_conn.commit()

sqlite_conn.close()
mdb_conn.close()

print("Carga de tablas completada")
```
